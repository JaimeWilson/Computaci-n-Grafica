```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib notebook
```

```python
def show_img(title,img):
    plt.figure(figsize = (8,6))
    plt.title(title)
    if img.ndim == 2:
        plt.imshow(img,cmap ='gray')
    else:
        plt.imshow(img)
    plt.show()
```
