```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.segmentation import mark_boundaries, slic
from mylib import *
%matplotlib notebook
```

```python
secao = np.load('secao_do_plug .npy')
show_gray(secao,"ORIGINAL")
```

```python
secao2 = secao[::2,::2]
```

```python
show_gray(secao2,"reduzida")
```

```python
mask = secao2>0
show_gray(mask,"mascara")
#show_hist(secao2.ravel(),"Histograma")
show_hist(secao2[mask].ravel(),"Histograma")
```

<hl> NORMALICE para ter um rango de 0 atÃ© 1 </hl>

```python
#perc = np.percentile(secao2[mask].ravel(),[0.1,99.9])
#show_npy(secao2[mask])
#perc
secao3 = secao2.astype(float)
vmax = np.amax(secao3)
vmin = np.amin(secao3)
secao3 = (secao3-vmin)/(vmax-vmin)
#show_hist(secao3.ravel(),"Normalizado")
show_hist(secao3[mask].ravel(),"Normalizado")
```

```python
segmentos = slic(secao3,n_segments=3000,compactness=0.1,max_iter=10,
                 sigma=0.1,slic_zero=False)

img_sp = mark_boundaries(secao3,segmentos,(1,0,0))


#mark_boundaries(image = secao3, label_img=mask, color: mask=mask, mode="'mask'")

show_rgb(img_sp,"Super Pixels")
```
