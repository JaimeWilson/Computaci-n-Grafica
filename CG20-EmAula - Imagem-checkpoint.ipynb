## CG20-Aula 4 - Imagem

```python
import matplotlib.pyplot as plt
import matplotlib.image as pim
import numpy as np
%matplotlib notebook
```

```python
img = pim.imread('Pilotis_PUC.jpg')
```

```python
def show_img(title,imajen):
    plt.figure(figsize = (8,6))
    plt.title(title)
    if imajen.ndim == 2:
        plt.imshow(imajen, cmap ='gray')
    else:
        plt.imshow(imajen)
    plt.show()
show_img("Pilotis", img)
```

```python
h, w = img.shape[:2]
print(img.shape, f'h={h}, w={w}')
```

```python
def lum(rgb):
    return np.dot(rgb,[0.2126, 0.7152, 0.0722])

gray = lum(img)
```

```python
print(gray.shape, img.shape)
```

```python
show_img("Pil", gray)
```

```python
x = np.arange(0,w,1)
y = gray[100,:]

plt.figure(figsize=(8,4))
plt.plot(x,y)
plt.show()
```
